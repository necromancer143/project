```python
#Create a weekly sales DataFrame
np.random.seed(0)
sales_amount = np.random.randint(20, size =(5,3))
weekly_sales = pd.DataFrame(sales_amount,
                            index = ["Mon", "Tues", "Wed", "Thus", "Fri"],
                            columns=["Almond Butter","Peanut Butter","Cashew Butter"])
weekly_sales
```

```python
#Price Table
prices= np.array([10, 8, 12])
prices
Butter_Prices = pd.DataFrame(prices.reshape(1,3),
             index = ["Prices"],
             columns=["Almond Butter","Peanut Butter","Cashew Butter"])
Butter_Prices
```

```python
#Total Amount Gained Every Week
Butter_Prices.shape, weekly_sales.shape
Total_Sales= Butter_Prices.dot(weekly_sales.T)
Total_Sales
```

```python
#Sum of all the amount gained in the week
Weekly_Amount=Total_Sales.values.sum()
Weekly_Amount
Total_Profit = pd.DataFrame(Weekly_Amount,
                            index=[""],
                            columns=["Sum_Amount"])
Total_Profit
```

```python
weekly_sales["Total($)"]=Total_Sales.T
weekly_sales
```
