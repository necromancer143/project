```python
#Price Table
prices= np.array([10, 8, 12])
prices
Butter_Prices = pd.DataFrame(prices.reshape(1,3),
             index = ["Prices"],
             columns=["Almond Butter","Peanut Butter","Cashew Butter"])
Butter_Prices
```

```python
#Total Amount Gained Every Week
Butter_Prices.shape, weekly_sales.shape
Total_Sales= Butter_Prices.dot(weekly_sales.T)
Total_Sales
```

```python
#Sum of all the amount gained in the week
Weekly_Amount=Total_Sales.values.sum()
Weekly_Amount
Total_Profit = pd.DataFrame(Weekly_Amount,
                            index=[""],
                            columns=["Sum_Amount"])
Total_Profit
```

```python
car_sales = pd.read_csv("/content/drive/MyDrive/car-sales-extended.csv")
car_sales.head()
```

```python
#Split into x/y
x=car_sales.drop("Price", axis = 1)
y=car_sales["Price"]


#split into training and test
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size = 0.2)
```

```python
#Turn the catergories into number
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer= ColumnTransformer([("one_hot",
                                 one_hot,
                                 categorical_features)],
                               remainder = "passthrough")
transformed_x = transformer.fit_transform(x)
transformed_x
```

```python
pd.DataFrame(transformed_x)
```

```python
#Fit the machine learning model
from sklearn.ensemble import RandomForestRegressor
np.random.seed(42)
x_train, x_test, y_train, y_test = train_test_split(transformed_x,
                                                    y,
                                                    test_size=0.2)
model = RandomForestRegressor()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
