```python
#inorder to fill the missing values in the table with the mean value
#DataType["Column name"].fillna(datatype["Column Name"].mean(), inplace= True)
#inplace is used to make the changed permanently to the table
#Bydefault inplace will be False
#dropna - remove all the NaN(Not available) information
neeknots
```

```python
#Creating a new column for the table
price = pd.Series([40,70,50,100,150,30,90,20,210])

#New column called Income
neeknots["Income_Per_CPC"] = price
neeknots
```

```python
neeknots.dropna(inplace=True)
neeknots
```

```python
#Column from Python List
#Length of the list should be same as the length of the rows available
Efficient = ["Y","N","Y","N","Y","N","Y","Y","Y"]
neeknots["Efficiency"]= Efficient
neeknots
```

```python
#Dropping a column from the table
neeknots = neeknots.drop("Previous position", axis=1)
neeknots
```

```python
#Use of Lambda
neeknots["Income_Per_CPC"] = neeknots["Income_Per_CPC"].apply(lambda x : x/10)
neeknots
```

```python
neeknots["Total Income"] = neeknots["CPC"]*neeknots["Income_Per_CPC"]
neeknots
```
