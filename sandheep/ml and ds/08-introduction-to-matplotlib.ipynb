## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for ploting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # width, heightb

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='simple plot', xlabel="x-axis", ylabel="y-axis")

# 5. Save and show 
fig.savefig("./matplortlib images/saved-sample-plot.png")
```

## Making figures with NumPy arrays

We want:

* Line plot
* Scatter plot 
* Bar plot

```python
import numpy as np
```

```python
# create a line plot
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# create scatter plot

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# making a plot from dictionary

nut_butter_prices = { "Almond butter": 10, "Peanut butter": 8, "Cashew butter": 12 }
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Nut Butter Store", xlabel="Items", ylabel="Price");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Nut Butter Store", ylabel="Items", xlabel="Price");
```

```python
# making some data for histogram and plot it

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(10000));
```

```python
# Option 2
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(10000));
```

## Which one should i use? (pyplot vs matplotlip object-oriented method?)
* When ploting something quickly, okay to use pyplot method
* when ploting something advanced, use the object oriented method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter", x='age', y='chol', c='target');
```

```python
# Object oriented method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
ax.set_xlim([45, 100]);
```

```python
# Object oriented method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# customize the graph
ax.set(title='Heart disease and chrolestrol level', xlabel='Age', ylabel='Chrolestrol level');

# adding a legend
ax.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10));

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

# customize ax0
ax0.set(title='Heart disease and choresterol level', xlabel='Age', ylabel='Chrosterol');

# adding legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax0.axhline(over_50['chol'].mean(), linestyle='--');


# Add data to ax1
scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])

# customize ax1
ax1.set(title='Heart disease and max heart rate', xlabel='Age', ylabel='Thalach');

# adding legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax1.axhline(over_50['thalach'].mean(), linestyle='--');
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

# customize ax0
ax0.set(title='Heart disease and choresterol level', ylabel='Chrosterol');

# adding legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax0.axhline(over_50['chol'].mean(), linestyle='--');


# Add data to ax1
scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])

# customize ax1
ax1.set(title='Heart disease and max heart rate', xlabel='Age', ylabel='Thalach');

# adding legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax1.axhline(over_50['thalach'].mean(), linestyle='--');
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'], cmap='winter')

# customize ax0
ax0.set(title='Heart disease and choresterol level', ylabel='Chrosterol');

# adding legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax0.axhline(over_50['chol'].mean(), linestyle='--');


# Add data to ax1
scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'], cmap='winter')

# customize ax1
ax1.set(title='Heart disease and max heart rate', xlabel='Age', ylabel='Thalach');

# adding legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax1.axhline(over_50['thalach'].mean(), linestyle='--');

# Adding title to the entire figure
fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold');
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10, 6))

# plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap='winter');

# customize the graph
ax.set(title='Heart disease and chrolestrol level', xlabel='Age', ylabel='Chrolestrol level');

# adding a legend
ax.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
plt.style.use('seaborn-whitegrid')
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'], cmap='winter')

# customize ax0
ax0.set(title='Heart disease and choresterol level', ylabel='Chrosterol');
ax0.set_xlim([50, 80])
# ax0.set_ylim([60, 200])

# adding legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax0.axhline(over_50['chol'].mean(), linestyle='--');


# Add data to ax1
scatter = ax1.scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'], cmap='winter')

# customize ax1
ax1.set(title='Heart disease and max heart rate', xlabel='Age', ylabel='Thalach');
ax1.set_ylim([60, 200])

# adding legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target');

# adding a horizontal line
ax1.axhline(over_50['thalach'].mean(), linestyle='--');

# Adding title to the entire figure
fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold');
```
